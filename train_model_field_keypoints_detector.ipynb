# Custom training

```python
import os

from ultralytics import YOLO


model = YOLO('yolov8n-pose.pt')  # load a pretrained model

model.train(data='/content/gdrive/My Drive/data_keypoints_detection_/config.yaml', batch=16, epochs=100, mosaic=0.0, plots=True, imgsz=640)
```
